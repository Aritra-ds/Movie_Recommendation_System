**id, title, genre, overview selecting those features for our project**

```python
movies = movies[['id', 'title', 'overview', 'genre']]
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['genre']
```

```python
movies
```

```python
new_data =  movies.drop(columns=['overview', 'genre'])
```

```python
new_data
```

Technique to convert the text dataframe into some back term
1. Bag of Words
2. TFIDF

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
new_data.info()
```

We want to calculate our distances based on the similarities

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vec:vec[1])
for i in distance[0:5]:
  print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies):
  index=new_data[new_data['title']==movies].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vec:vec[1])
  for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
recommand("Iron Man")
```
